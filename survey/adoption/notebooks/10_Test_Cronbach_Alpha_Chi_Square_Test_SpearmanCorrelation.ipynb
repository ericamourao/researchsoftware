Cronbach’s alpha is the most used indicator of scale reliability

Internal Consistency = 0.9 ≤ α Excellent, = 0.8 ≤ α < 0.9 Good, = 0.7 ≤ α < 0.8 Acceptable, = 0.6 ≤ α < 0.7 Questionable, 0.5 ≤ α < 0.6 Poor



**Result: Years of Experience versus profile**

The first value (3.4700055374743437) is the Chi-square value, followed by the p-value (0.6279309539437089),

Since all of the expected most of frequencies are greater than 5, the chi2 test results can be trusted.

**3) Correlation**

Spearman Rank Correlation Coefficient

Correlation refers to the association between the observed values of two variables.

The value of a correlation coefficient can range from -1 to 1, with the interpretations:

-1: is a perfect negative relationship between two variables

0: is no relationship between two variables

1: is a perfect positive relationship between two variables

Spearman Rank Correlation is used to measure the correlation between two ranked variables.

**3.2) Heat Map of Correlation**

```python
def display_correlation(df):
    r = df.corr(method="spearman")
    plt.figure(figsize=(60,30))
    heatmap = sns.heatmap(df.corr(method="spearman"), vmin=-1, 
                      vmax=1, annot=True, fmt=".3f",  linewidth=.1, annot_kws={"size":18})
    plt.title("Spearman Correlation")
    #heatmap.xaxis.tick_top()
    #sns.set(font_scale=2) 
    return(r)
```
